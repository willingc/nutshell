# Compose a request to discourse

````python
import os

# nicer display of json
from pprint import pprint

# queries
import requests
````

````python
# Utility


def pprint_json(jstr):
    from json import dumps, loads

    print(dumps(loads(jstr), indent=2))
````

````python
# Auth
token = os.environ.get("DISCOURSE_API_KEY")

headers = {"Authorization": f"Bearer {token}", "Content-Type": "application/json"}
````

````python
# get a topic
response = requests.get(
    "https://discuss.python.org/t/104906.json?print=true", headers=headers
)
````

````python
# Show status code, content-type, and JSON body
print(response.status_code, response.headers["Content-Type"])
response.text
````
